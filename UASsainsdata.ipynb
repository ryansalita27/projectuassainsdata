```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DataCovid.csv')
df
```

```python
top10 = df.head(10)
top10
```

```python
x = top10.Provinsi
y1 = top10.Dirawat
y2 = top10.Terkonfirmasi
y3 = top10.Sembuh
y4 = top10.Meninggal
```

```python
plt.plot(x,y1)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(x, y1)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(x, y1, label='Dirawat')
plt.plot(x, y2, label='Terkonfirmasi')
plt.plot(x, y3, label='Sembuh')
plt.plot(x, y4, label='Meninggal')
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.show()
```

```python
df2 = pd.DataFrame([y1, y2, y3, y4])
df2.plot.bar()
plt.legend(x, loc='upper left')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.bar(stacked=True);
plt.legend(x, loc='upper left')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.barh()
plt.legend(x, loc='upper right')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.barh(stacked=True)
plt.legend(x, loc='upper right')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
df2.plot.area()
plt.legend(x, loc='upper left')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
factor=0.0009
labels = ['Dirawat', 'Terkonfirmasi', 'Sembuh', 'Meninggal']
plt.scatter(x, y1, s=y1*factor)
plt.scatter(x, y2, s=y2*factor);
plt.scatter(x, y3, s=y3*factor);
plt.scatter(x, y4, s=y4*factor);
plt.legend(labels, loc='upper right')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.pie(y1, labels=x)
plt.title('Kasus Dirawat')
plt.show()
```

```python
plt.pie(y2, labels=x)
plt.title('Kasus Terkonfirmasi')
plt.show()
```

```python
plt.pie(y3, labels=x)
plt.title('Kasus Sembuh')
plt.show()
```
